```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# loading Dataset
df=pd.read_csv('/content/insurance.csv')
df.head()
```

```python
df.shape
```

```python
# checking null values
df.isnull().sum()
```

```python
# Converting all Non-numerical  variables to numeric
Sex={'male':0,'female':1}
Smoker={'yes':1,'no':0}
Region={'southwest':0,'southeast':1,'northeast':2,'northwest':3}
```

```python
# Mapping the values to  the Dataset
df['sex']=df['sex'].map(Sex)
df['smoker']=df['smoker'].map(Smoker)
df['region']=df['region'].map(Region)
```

```python
# Visualizing the New Data
df.head()
```

```python
# IdepedentVariables
# age
# bmi
# sex
# smoker
# number of children
# Region

# Target Variable
# Charges
# Create subplots for all Target Variable vs All the 6 indepedent Variables
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

# Scatter plot for variable1
axs[0, 0].scatter(df['age'],df['charges'] )
axs[0, 0].set_title('Age')

# Scatter plot for variable2
axs[0, 1].scatter(df['bmi'],df['charges'] )
axs[0, 1].set_title('BMI')

# Scatter plot for variable3
axs[0, 2].scatter(df['sex'],df['charges'] )
axs[0, 2].set_title('Sex')

# Scatter plot for variable4
axs[1, 0].scatter(df['children'],df['charges'] )
axs[1, 0].set_title('Children')

# Scatter plot for variable5
axs[1, 1].scatter(df['smoker'],df['charges'] )
axs[1, 1].set_title('Smoker')

# Scatter plot for variable6
axs[1, 2].scatter(df['region'],df['charges'] )
axs[1, 2].set_title('Region')

# Adjust layout
plt.tight_layout()
plt.show()
```
